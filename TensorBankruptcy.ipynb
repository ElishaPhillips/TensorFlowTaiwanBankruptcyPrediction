```python
import os 
import pandas as pd
import numpy as np


from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential, load_model


import tensorflow as tf
from tensorflow.keras.layers import Dense, Flatten, Conv2D
from tensorflow.keras import Model
from sklearn.metrics import accuracy_score

os.chdir('C:\\Users\\Elisha\\Documents\\Python\\Portfolio\\ML\Bankruptcy')
```

```python
bankruptcy = pd.read_csv('data.csv')

bankruptcyi = bankruptcy.rename(columns={x:y for x,y in zip(bankruptcy.columns,range(0,len(bankruptcy.columns)))})

bankruptcyi.head()
```

```python
X = pd.get_dummies(bankruptcyi.drop(0, axis=1))
y = bankruptcyi[0]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2)
```

```python
model = Sequential()
model.add(Dense(units=64, activation='relu', input_dim=len(X_train.columns)))
model.add(Dense(units=128, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='sgd', metrics='accuracy')
```

```python
model.fit(X_train, y_train, epochs=500, batch_size=32)
```

```python
y_hat = model.predict(X)
y_hat = [1 if val < 0.5 else 0 for val in y_hat]

accuracy_score(y, y_hat)
```
